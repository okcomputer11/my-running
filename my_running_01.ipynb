```python
############################################################################
# now I want to calculate the linear regression "by hand"
#
# x = x_training --> Pace_in_seconds
# y = y_training --> Distanz
#
##################################
x_cuadrado = x_training**2
x_y = x_training * y_training
##################################
sum_x_training = sum(x_training)
print(f'Sum of Pace (in seconds): {sum_x_training}')
sum_y_training = sum(y_training)
print(f'Sum of Distanz (in km): {sum_y_training}')
sum_x_cuadrado = sum(x_cuadrado)
sum_x_y = sum(x_y)
##################################
# y = ax + b
#
#     n sum(xy) - sum(x)sum(y)
# a = ---------------------------
#     n sum(x^2) - sum(x)^2
#
#
n = len(dfrunning)
a = ((n*sum_x_y)-(sum_x_training*sum_y_training)) / ((n*sum_x_cuadrado) - (sum_x_training**2))
#
#     sum(y) - a sum(x)
# b = -----------------
#            n
#
#
b = (sum_y_training - a*sum_x_training) / n

print(f'total de filas: {n}')
print('formula recta regresion lineal  y = ax + b')
print(f'a: {a}')
print(f'b: {b}')
print('------------------------')
print(f'|  y = {float(a):.2f}x + {float(b):.2f}  |')
print('------------------------')
print()

y_resultant = []
for x in [340, 240, 260]: #running.Pace_in_seconds:
    punto = ( (float(a) * float(x)) + float(b))
    print(f'x: {x} --> y: {punto:.2f}')
    y_resultant.append(punto)
    
# it WORKS !!!!
# the points of 'y_resultant' pass through the line calculated in the linear regression
```

```python
# Korrelationskoeffizient zwischen Pace und Distanz
correlation = dfrunning['Pace_in_seconds'].corr(dfrunning['Distanz'])
print("Korrelationskoeffizient zwischen Pace_in_seconds und Distanz:", correlation)

#################################################################################################
# JETZT I want to calculate the Korrelationskoeffizient zwischen Pace und Distanz ...  by hand ##
x_mean = dfrunning.Pace_in_seconds.mean()
x_mean2 = sum(dfrunning.Pace_in_seconds) / n
print(f'mean of Pace: {x_mean} = {x_mean2}')
y_mean = dfrunning.Distanz.mean()
print(f'mean of Distanz: {y_mean:.2f}\n')

# ∑(x - ~x)
x_minus_xmean = []
print('    genero arreglo de (x - ~x) Pace in seconds')
for hx in range(len(dfrunning)): 
    vx = dfrunning.Pace_in_seconds[hx] - x_mean
    print(f"{hx}){vx:.2f}", end='  ')
    x_minus_xmean.append(vx) # genero arreglo de (x - ~x)
print(f'\n∑(x - ~x): {sum(x_minus_xmean):.2f}\n\n') # <-- la suma debe dar SIEMPRE 0 !!

# ∑(y - ~y)
y_minus_ymean = []
print('    genero arreglo de (y - ~y) Distanz')
for hy in range(len(dfrunning)):
    vy = dfrunning.Distanz[hy] - y_mean
    print(f"{hy}){vy:.2f}", end='  ')
    y_minus_ymean.append(vy) # genero arreglo de (y - ~y)
print(f'\n∑(y - ~y): {sum(y_minus_ymean):.2f}\n\n') # <-- la suma debe dar SIEMPRE 0 !!

############################################################################################

# ∑(x - ~x)^2 
cuadratic_of_x_minus_xmean = []
print('    genero arreglo de (x - ~x)^2 ')
for tx in range(len(dfrunning)):
    cuadratic_of_x_minus_xmean.append(x_minus_xmean[tx]**2) # genero arreglo de (x - ~x)^2
    print(f"{tx}){x_minus_xmean[tx]**2:.2f}", end='  ')
print(f'\n∑(x - ~x)^2: {sum(cuadratic_of_x_minus_xmean):.2f}\n\n')

# ∑(y - ~y)^2 
cuadratic_of_y_minus_ymean = []
print('    genero arreglo de (y - ~y)^2 ')
for ty in range(len(dfrunning)):
    cy = y_minus_ymean[ty] ** 2
    cuadratic_of_y_minus_ymean.append(cy) # genero arreglo de (y - ~y)^2
    print(f"{ty}){cy:.2f}", end='  ')
print(f'\n∑(y - ~y)^2: {sum(cuadratic_of_y_minus_ymean):.2f}\n\n')

############################################################################################

# ∑(x - ~x)(y - ~y) 
x_xmean_mal_y_ymean = []
print('    genero arreglo de (x - ~x)(y - ~y) ')
for f in range(len(dfrunning)):
    s = x_minus_xmean[f] * y_minus_ymean[f]
    x_xmean_mal_y_ymean.append(s) # genero arreglo de (x - ~x)(y - ~y)
    print(f"{f}){s:.2f}", end='  ')
print(f'\n∑(x - ~x)(y - ~y): {sum(x_xmean_mal_y_ymean):.2f}\n\n')

############################################################################################

# sqrt ( ∑(x - ~x)^2 )
ba = (sum(cuadratic_of_x_minus_xmean))**0.5
print(f'sqrt ( ∑(x - ~x)^2 ) -- > {ba}')
print()

# sqrt ( ∑(y - ~y)^2 )
bb = (sum(cuadratic_of_y_minus_ymean))**0.5
print(f'sqrt ( ∑(y - ~y)^2 ) -- > {bb}')
print()

r = sum(x_xmean_mal_y_ymean) / ( ba * bb ) 

print(f'\nlinear correlation coefficient (by hand): {r:.2f}\n\n')
```
